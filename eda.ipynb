```python
import pandas as pd
```

Загрузка

```python
X_train = pd.read_csv('data/X_train.csv', sep=';')
X_test  = pd.read_csv('data/X_test.csv',  sep=';')
y_train = pd.read_csv('data/y_train.csv', sep=';')
qual    = pd.read_csv('data/pk-pp.csv',   sep=';')
school  = pd.read_csv('data/school.csv',  sep=';')
edu     = pd.read_csv('data/edu.csv',     sep=';')
```

Анализ пропусков

```python
datasets = [X_train, X_test, y_train, qual, school, edu]
all_rows, all_skips = 0, 0
for dataset in datasets:
    if dataset.size != 0:
        print(f'Процент пропусков = {dataset.isna().sum().sum() / (dataset.size):.2f}%')
    all_rows += dataset.size
    all_skips += dataset.isna().sum().sum()

print('\n')
print(f'Общий процент пропусков = {all_skips / all_rows:.2f}%')
for dataset in datasets:
    dataset.fillna()
```
